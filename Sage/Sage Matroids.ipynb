```sagemath
n = 9
r = 4
```

<span style='font-size:large'>This notebook includes the function "filter\_matroids" which creates a file of filtered matroids of rank \$r\$ on \$n\$ ground set elements according to desired matroid properties. Some optional properties include: </span>

M.is\_simple\(\) \-\- returns TRUE if M is simple 

M.is\_cosimple\(\) \-\- returns TRUE if M is cosimple 

M.is\_binary\(\) \-\- returns TRUE if M is binary 

M.is\_ternary\(\) \-\- returns TRUE if M is ternary 

M.is\_regular\(\) \-\- returns TRUE if M is regular 

M.is\_graphic\(\) \-\- returns TRUE if M is graphic

not M.loops\(\) \-\- returns TRUE if M is loopless



```sagemath
# Create matroid iterator
matroid_iterator = [M for M in AllMatroids(n, r) if not M.loops() and M.is_ternary()]

# Create appropriate file name
filename = "all_matroids94_0_ternaryloopless_bases.txt"
```

```sagemath
def filter_matroids(n, r, filename):
    # Open the file and write each matroid (with bases) satisfying desired properties
    with open(filename, 'w') as f:
        for i, M in enumerate(matroid_iterator):
            f.write(f"# MATROID {i}\n")
            for basis in M.bases():
                line = ' '.join(str(elem) for elem in basis) + '\n'
                f.write(line)
            f.write('\n')
    print(f"Successfully wrote all bases to {filename}")
```

```sagemath
filter_matroids(n,r,filename)
```
